```python
# 3. Creating Dictonary
states_to_region={
    'Northeast' : ['Maine','New Hampshire','Vermont','Massachusetts','Rhode Island','Connecticut','New York','Pennsylvania','New Jersey'],
    'Midwest' : ['Wisconsin',' Michigan','Illinois','Indiana','Ohio','North Dakota','South Dakota','Nebraska','Kansas','Minnesota','Iowa','Missouri'],
    'South' : ['Delaware','Maryland','District of Columbia','Virginia','West Virginia','North Carolina','South Carolina','Georgia','Florida','Kentucky','Tennessee','Mississippi','Alabama','Oklahoma','Texas','Arkansas','Louisiana'],
    'West' : ['Idaho','Montana','Wyoming','Nevada','Utah','Colorado','Arizona','New Mexico','Alaska','Washington','Oregon','California','Hawaii']
    }
```

```python
# Create a function to map states to region
def map_states_to_region(state):
    for region, states in states_to_region.items():
        if state in states:
            return region
    return 'Unknown'
```

```python
# Define the predict_profile function
def profile(row):
    age = row['Age']
    fam_status = row['fam_status']
    n_dependant = row['n_dependants']

    if age >= 18 and age <= 40:
        if fam_status == 'single' and n_dependant == 0:
            return 'Single Adult'
        elif fam_status == 'married' and n_dependant >= 2:
            return 'Young Parent'
        elif fam_status == 'divorced/widowed' and n_dependant >= 1:
            return 'Single Parent'
        elif fam_status == 'divorced/widowed' and n_dependant == 0:
            return 'Undefined'
        else:
            return 'Living with parents'
    else:
        return 'Undefined'
```

```python
 #ords_prods_group[ords_prods_group['fam_status']=='divorced/widowed'][['department_id','n_dependants']]
```

```python
# Apply the function
ords_prods_group_dept = ords_prods_group[ords_prods_group['department_id']==16]
ords_prods_group_profile = ords_prods_group_dept[['user_id','Age','n_dependants', 'fam_status', 'income', 'product_name', 'department_id','prices', 'order_id', 'orders_day_of_week', 'order_hour_of_day']]

ords_prods_group_profile['fam_profile'] = ords_prods_group_dept.apply(profile, axis = 1)
```
